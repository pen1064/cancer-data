```python
df = pd.read_csv(r'C:\Users\Penelope\Desktop\ML\practice_coding_challenge\breast_cancer_wisconsin\data.csv')
y = df['diagnosis']
df.drop(columns=['diagnosis'], inplace=True)
y = y.replace('M',0)
y = y.replace('B', 1)
print(Counter(y))
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(df, y, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

class ColumnDrop:
    def __init__(self, columns):
        self.columns = columns
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X.drop(self.columns, axis=1)

num_pipe = Pipeline([('drop', ColumnDrop(['id'])),('scale', MinMaxScaler())])
num_pipe.fit_transform(Xtrain)
```

```python
from collections import Counter
Counter(ytrain) 
```

not so balance 

```python
import xgboost
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score
pos_weight = np.sum(ytrain==0)/np.sum(ytrain==1)
model_pipe = Pipeline([('clf', XGBClassifier(n_jobs=-1, random_state=0, scale_pos_weight=pos_weight))])
overall_pipe = Pipeline([('clean', num_pipe), ('model', model_pipe)])
overall_pipe.fit(Xtrain, ytrain)
y_pred = overall_pipe.predict(Xtest)
print(roc_auc_score(y_pred, ytest))
print(classification_report(y_pred, ytest))
```

```python
import joblib
import pickle
joblib.dump(overall_pipe, 'overall_pipe.joblib', protocol=pickle.HIGHEST_PROTOCOL)
```

```python
from hyperopt import fmin, Trials, tpe, STATUS_OK, hp
from xgboost import DMatrix
import xgboost 
from sklearn.model_selection import KFold
from functools import partial
SPACE = {
    'params':{
        'max_depth': hp.quniform('max_depth', 3, 8, 1),
        'min_child_weight': hp.quniform('min_child_weight', 1, 12, 2),
        'subsample': hp.uniform('subsample', 0.6, 1),
        'reg_alpha': hp.uniform('reg_alpha', 0.1, 10),
        'reg_lambda':hp.uniform('reg_lambbda', 0.1, 10),
        'gamma':  hp.uniform('gamma', 0, 1),
        'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.1))
    }
}


def objective(hps, X, y, pipe):
    hps['params']['max_depth'] = int(hps['params']['max_depth'])
    hps['params']['min_child_weight'] = int(hps['params']['min_child_weight'])

    XG = DMatrix(pipe.fit_transform(X),y)
    result = xgboost.cv(hps['params'], XG, num_boost_round = 100, nfold=4, stratified=False, metrics='auc')
    score = np.mean(result['test-auc-mean'])
    loss = 1-score #trying to maximize the score, that is why have to flip sign
    return {'loss': loss, 'params': hps, 'status': STATUS_OK}

test_trial = Trials()

bestp = fmin(partial(objective, X = Xtrain, y = ytrain, pipe = num_pipe), SPACE, algo=tpe.suggest, max_evals=100, trials=test_trial)
```
